```python
from PIL import Image
from pycocotools.coco import COCO

import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, '../libs')

from imantics import Image, BBox, Mask, Polygons
```

```python
image = Image.from_path("data/tesla.jpg")
image.add(Mask(np.random.choice([0, 1], size=(600, 900), p=[49./50, 1./50])))
image.add(Polygons([[100, 200, 500, 100, 700, 500, 500, 500]]))
image.add(BBox((20, 50, 100, 200)))
```

```python
plt.imshow(image.array)
plt.title("Image")
plt.show()

plt.imshow(image.draw_masks())
plt.title("Draw masks of annotations")
plt.show()

plt.imshow(image.draw_bboxs())
plt.title("Draw bboxs of annotations")
plt.show()
```
