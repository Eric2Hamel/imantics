```python
from PIL import Image
from pycocotools.coco import COCO

import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, '../libs')

import imglang as il
```

```python
# Set segment
annotation = il.Annotation(width=10, height=10, segments=[[3, 9, 7, 8, 1, 2]])

mask = annotation.mask
bbox = annotation.bbox

mask[bbox[1], bbox[0]] = 2
mask[bbox[3], bbox[2]] = 2

plt.imshow(annotation.mask)
```
